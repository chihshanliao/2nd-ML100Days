```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
#print(f'{len(num_features)} Numeric Features : {num_features}\n')
print(len(num_features), 'Numeric Features :', num_features, '\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

* 確實可以進一步提高分數(如下列數據所示)

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
# 將 1stFlrSF 限制在 450 到 2400 以內, 調整離群值
#df['1stFlrSF-2'] = df['1stFlrSF'].clip(500, 2300) #0.8822761099448131
#df['1stFlrSF-2'] = df['1stFlrSF'].clip(450, 2350) #0.887032127713046
#df['1stFlrSF-2'] = df['1stFlrSF'].clip(450, 2400) #0.8877045777039726
#df['1stFlrSF-2'] = df['1stFlrSF'].clip(450, 2450) #0.8879677686089764
df['1stFlrSF-2'] = df['1stFlrSF'].clip(450, 2000) #0.8776733463127849

sns.regplot(x = df['1stFlrSF-2'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

* 可以再進一步提高分數
  但捨棄離群資料時, 因為也同時減少了驗證的資料, 因此捨棄離群值未必比調整離群值好
  實際結果仍需再驗證

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
# 將 1stFlrSF 限制在 450 到 2400 以內, 捨棄離群值
#keep_indexs = (df['1stFlrSF']> 450) & (df['1stFlrSF']< 2400) #0.8776733463127849
#keep_indexs = (df['1stFlrSF']> 500) & (df['1stFlrSF']< 2350) #0.8776733463127849
#keep_indexs = (df['1stFlrSF']> 600) & (df['1stFlrSF']< 2350) #0.8776733463127849
#keep_indexs = (df['1stFlrSF']> 600) & (df['1stFlrSF']< 1900) #0.8776733463127849
keep_indexs = (df['1stFlrSF']> 450) & (df['1stFlrSF']< 2000) #0.8776733463127849

df2 = df[keep_indexs]
train_Y2 = train_Y[keep_indexs]
sns.regplot(x = df2['1stFlrSF'], y=train_Y2)
plt.show()

# 做線性迴歸, 觀察分數
train_X2 = MMEncoder.fit_transform(df2)
estimator = LinearRegression()
cross_val_score(estimator, train_X2, train_Y2, cv=5).mean()
```
